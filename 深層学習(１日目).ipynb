# 〇識別と生成

## １．識別モデル

識別モデルとは、データを目的のクラスに分類することである。  
識別モデルの基本は、データ量の多いものから、データ量の少ないものを生成する。  
> (例)：画像を犬や猫と識別する(画像を入力とし、クラスの確率を出力する)  
>> データ：画像  
>> クラス：犬＆猫
 
「あるデータ$x$が与えられた」という条件において、クラス$C_{k}$が成立する確率($P$)($k$：種類)  
> ### $P(C_{k}|x)$  

### 1.1. 識別モデル(具体例)  
> (1) 決定木  
> (2) ロジスティック回帰  
> (3) サポートベクターマシン(SVM)  
> (4) ニューラルネットワーク  


## ２．生成モデル

生成モデルとは、特定のクラスのデータを生成することである。(⇔ 識別モデル)  
生成モデルの基本は、データ量の少ないものから、データ量の多いものを生成するモデルである。  

> (例)：犬の画像がどのようなものかを学習して、犬の画像を生成する(クラスを入力として、画像を出力とする)  
>> 特定のクラス：犬(１種類)  
>> データ：犬っぽい画像

「あるクラス$y$」に属するという条件において、データ$(C_{k}$が分布する確率$P$  
> ### $P(x|C_{k})$  

犬・猫などのクラス(種類)のように低次元データから、画像のように高次元なデータを出力するため、必要な学習データは識別モデルより多く必要である。学習データが多くなると、必要な計算量や学習時間が大きくなる。    

### ２.1. モデル(具体例)   
> (1) 隠れマルコフモデル  
> (2) ベイジアンネットワーク  
> (3) 変分オートエンコーダ(VAE)  
> (4) 敵対的生成ネットワーク(GAN)  


## ３．識別関数

入力値$x$を直接クラスに写像(変換)する関数$f(x)$を推定する。  
各クラスの確率を出すのではなく、どのクラスに属するのかのみを出力するため、学習量は少なくて済む。

# 〇活性化関数

活性化関数とは、ニューラルネットワークにおいて、次の層への出力の大きさを決める”非線形”の関数である。  
次の層への値を伝える/伝えない、またはその強弱を設定する。  

## １．中間層に用いられる活性化関数

### 1.1. ステップ関数

ステップ関数とは、閾値を超えたら発生する関数である。  
ステップ関数は、線形分離が可能なものしか学習することができないため、現在は活性化関数として殆ど使用されていない。  
#### ${f(x) = \begin{cases}
1 \ \ (x \ge 0) \\
0 \ \ (x < 0) \end{cases}}$  
> 出力値："０"or"１"のみ  


```python
# ステップ関数を実装する(p45)
def step_function(x):
    if x > 1:
        return 1
    else:
        return 0
```

### 1.2. シグモイド関数

シグモイド関数とは、"0"から"1"の間を緩やかに変化する関数である。  
シグモイド関数は"0"から"1"の間を緩やかに変化する関数である故に、微分を使用した学習が上手く進めることができる。  
その一方で、大きい値になると微分による変化が微小になるため、"勾配消失問題"が発生する。  
### $f(x) = \frac{1}{1 + e^{-x}}$  


```python
# ステップ関数を実装する(p48)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### 1.3. ReLU関数(正規化線形関数_(Reflected Linear Unit))

ReLU関数とは、最も使用されている関数である。  
ReLU関数では、入力($x$)が"0"より大きければそのまま出力し、入力($x$)が"0"より小さければ"0"と出力する。  
ReLU関数を使用することによって、"勾配消失問題"の回避や"スパース化"に貢献することができる。  
> (例)：画像の畳み込みの場合、RGBは"0~255"の正の値しか存在しない。しかし、重み($w$)とバイアス($b$)を基に算出した値が負の値である場合、不要なデータとして画像の特徴を捉えることができない。その対策として、ReLU関数を使用することによって、負の値を"0"として処理する。  

### $f(x) = \begin{cases}
x \ \ (x > 0) \\
0 \ \ (x \le 0) \end{cases}$

```python
# ReLU関数(正規化線形関数)を実装する(p52)
def relu(x):
    return np.maximum(0, x)
```

## 1.4. その他の活性化関数

### 1.4.1. tanh(双曲線関数)  
> "-1"から"1"の間で出力する  

```python
# tanh(双曲線関数)を実装する(p183)  
def tanh(x):
    return np.tanh(x)
```

### 1.4.2. LReLU("ReLU"関数の拡張(Leaky Rectified Linear Unit))  
> 出力値が"0"以下の場合、入力値を$α$倍した値を出力する($α$：基本的に0.01(固定))    

```python
# tanh(双曲線関数)を実装する([引用文献] https://qr.paps.jp/Lh3cb)  
def lrelu(x, alpha=0.01):
    return np.where(x >= 0.0, x, alpha * x)
```

### 1.4.3. PReLU("LReLU"関数の拡張) 
> 出力値が"0"以下の場合、入力値を$α$倍した値を出力する($α$：学習によって動的に決まる)  

(参照文献)：https://qr.paps.jp/jHwqs  

### 1.4.4. ELU("LReLU"関数や"PReLU"関数を改良した関数)  
> 出力値が$0$以下の場合、"0"から"α"(基本的に"1.0")の値を出力する(オイラー数を底とする指数関数を使用する)  

(参照文献)：https://qr.paps.jp/o69n7  

```python
# "ELU"を実装する([引用文献] https://qr.paps.jp/o69n7)  
def elu(x, alpha=1.0):
    return np.where(x > 0.0, x, alpha * (np.exp(x) - 1))
```

### 1.4.5. SELU("ELU"関数の拡張)  
> 入力値が"0"以下でも"0"より大きい場合でも、"λ"倍した値を出力する  

(参照文献)：https://qr.paps.jp/cFFU1

```python
# "SELU"を実装する([引用文献] https://qr.paps.jp/cFFU1)  

SCALE = 1.0507009873554804934193349852946
ALPHA = 1.6732632423543772848170429916717

def selu(x, scale=SCALE, alpha=ALPHA):
    return scale * np.where(x > 0.0, x, alpha * (np.exp(x) - 1))
```

### 1.4.6. Swish(=SiLU)  
> 入力値が"0"以下の場合、ほぼ"0"の値を出力する  
> 入力値が"0"以上の場合、ほぼ入力値そのままを出力する

# 〇出力層

分類は、人間が欲しい最終的な結果を出力する。(各クラスに属する確率等)  
ニューラルネットワークの学習において、(1)入力データと(2)訓練データの２つを準備する。  
そして"出力層の結果"と"目的変数の正解値"を比較して、どの程度合っている(ズレている)のかを算出するために"誤差関数"を利用する。  

## １．誤差関数

分類の場合、"各クラスが属する確率"と"One-hotベクトル"で表現されている各クラスの正解値"をもとに、どれだけズレているのかを算出する。  
誤差関数は、ズレているほど値は大きくなる。  

<用語>  
One-hotベクトル：正解ラベルを"1"として、それ以外は"0"で表現する表記法のこと  

### 1.1. 平均二乗誤差(MSE：Mean squared Error)

### $MSE = \frac{1}{2}\sum_{k} { ( y_{k} - t_{k}) }^2$

```python
# 平均二乗誤差の関数実装(p89)
def sum_square_error(y, t):
    return 0.5 * np.sum((y-t)**2)
```

### 1.2. 交差エントロピー誤差(分類問題の誤差関数)

### $E = -\sum_{i=k}{t_{k}log\ y_{k}}$  

> $t_{k}$：正解ラベル(One-Hot表現)  
> $log$：底が$e$の自然関数($log_{e}$)  
> $y_{k}$：ニューラルネットワークの出力  

```python
# 交差エントロピー誤差の実装(p91)
def cross_entropy_error(y, t):
    delta = 1e-7 #微小な値(マイナス無限大を発生防止のため)
    return -np.sum(t * np.log(y + delta))
```

### 2.1. シグモイド関数

### $f(x) = \frac{1}{1 + \exp(-x)}$  
> ("1" のクラスに該当する確率)と("0" のクラスに該当する確率)の合計値は"1"となる  

二値分類で１クラスに該当する確率を出力する(= ロジスティック回帰において同じように使用する)   

```python
# シグモイド関数の実装(p48)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### 2.2. ソフトマックス関数

### ${y_{k} = \frac{e^{a_{k}}}{\sum_{i=1}^{n}{e^{a_{i}}}}}$  
> 分子$({e^{a_{k}}})$：入力値の指数関数  
> 分母$({e^{a_{i}}})$：全ての入力値の指数関数の和("1")  

他クラス分類(3クラス以上)において使用する。  
各クラスに該当する確率を出力する。  
クラス数($K$)ごとに全ての確率を合計すると"1"になる。  


```python
# ソフトマックス関数を定義する(p67)
# -------------------------------------------
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y
# -------------------------------------------
import numpy as np
a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a) #指数関数
print('〇指数関数の値：', exp_a)

sum_exp_a = np.sum(exp_a) #指数関数の合計
print('〇指数関数の合計：', sum_exp_a)

y = exp_a / sum_exp_a
print('〇ソフトマックス関数', y)
# -------------------------------------------
```

ソフトマックス関数(オーバーフロー対策)➡入力値をの中で最大値を採用すること

```python
# ソフトマックス関数(オーバーフロー対策)を定義する(p67)
# -------------------------------------------
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y
# -------------------------------------------
import numpy as np
a = np.array([1010, 1000, 990]) #ソフトマックス関数の計算※正しく計算できない
c = np.max(a) #入力値の中で最大値を採用する
print(a-c)
np.exp(a - c) / np.sum(np.exp(a - c))
# -------------------------------------------
```

# 〇勾配降下法

### １．勾配降下法  
勾配降下法とは、ニューラルネットワークを学習させる手法である  

### $w^{(t+1)} = w^{(t)} - \varepsilon \nabla E$  

> $w^{(t)}$：前回の学習  
> $\varepsilon \nabla E$：間違えた部分  
>> $\nabla E = \frac{\partial E}{\partial w} = \left[\frac{\partial E}{\partial w_{1}}, \cdots, \frac{\partial E}{\partial w_{M}} \right]$
 
＜勾配降下法：学習＞  
> ・学習は何回も繰り返し行う  
> ・前回の学習の答えから間違えた部分を引く(調整して次回の学習に活用する)  

＜勾配降下法の種類＞  
> (1) 勾配降下法(バッチ勾配降下法)   
> (2) 確率的勾配降下法(SGD)  
> (3) ミニバッチ勾配降下法  

## 2. 学習率

学習率$(η)$とは、パラメータの収束に関わるハイパーパラメータである。(勾配をどれだけの勢いで降りるかというイメージ)  

・学習率$(η)$が大きすぎる場合  
> 最適解(最小値)を飛び越えて値が大きくなり、学習がうまく進まない(分散)  

・学習率$(η)$が小さすぎる場合
> 最適解(最小値)に辿り着くまでに何度も更新するため、局所最適解(極小値)を計算してしまう可能性がある  


### 2.1．学習率$(η)$を決定する手法

#### 〇 Momentum(モーメンタム)  
> 勾配の移動平均を出して振動を抑える(過去の勾配を考慮することによって、急激な変化を抑える)  
#### ・ネステロフの加速勾配法(NAG)  
> 勾配が正しい向き(損失を減らす方向)に向いていることを保証したMomentum(モーメンタム)  
#### 〇 AdaGrad  
> 次元ごとに学習率を変化させるようにしたもの  
#### 〇 RMSprop("AdaGrad"の改良版)  
> 学習率が"0"に１度でも近づくとほどんど変化しなくなる  
#### ・AdaDelta("RMSprop"の改良版)  
> 次元数のミスマッチを解消したもの  
#### ◎ Adam(最も使用されている)  
> 移動平均と学習率の調整(Momentum + RMSprop)

## ３. エポック  

エポックとは、誤差関数の値を小さくする方向に重み($W$)とバイアス($b$)を更新し、次回の学習に反映することである  
> (例)：全てのデータセットを１度与える場合："１エポック"  
> (例)：同じデータセットを2回与える場合："２エポック"　　

## ４．オンライン学習 / バッチ学習  

### 4.1．オンライン学習  
オンライン学習とは、データをその都度与えてから学習させる学習方法である(⇔ バッチ学習)   
オンライン学習では、リアルタイムで手に入るデータについて、対応した学習を迅速にすることができるので利用しやすい  

### 4.2．バッチ学習  
バッチ学習は、データを１通り用意してから１度に与えて学習させる学習手法である  
バッチ学習の場合、１度に大量のデータをメモリ上に展開しないといけない  

## ５．確率的勾配降下法  

### $w^{(t+1)} = w^{(t)} - \varepsilon \nabla E$  

・データが冗長な場合、計算コストを削減することができる  
・毎回異なるデータを用いて学習をするため、局所最適解に収束するリスクを軽減することができる  
・オンライン学習の手法を利用することができる  

## ６．ミニバッチ勾配降下法  

### $w^{(t+1)} = w^{(t)} - \varepsilon \nabla E$  

・オンライン学習の派生版(オンライン学習の利点をバッチ学習に導入したもの)  
・ミニバッチとは、小分けのことである  
・確率的勾配降下法のメリットを維持しながら、計算処理を複数のコンピュータが並列して処理を行うことで計算コストを削減する

# 〇誤差逆伝播法

### $\nabla E = \frac{\partial E}{\partial w} = \left[\frac{\partial E}{\partial w_{1}}, \cdots, \frac{\partial E}{\partial w_{M}}\right]$

誤差逆伝播法とは、出力層側から微分を用いて誤差を算出して前の中間層へと伝播することによって、各パラメータの微分値を最小限の計算で求める分析手法である。  
➡ 順伝播によって算出した誤差を微分して前の層に戻すことによって、重み($W$)とバイアス($b$)を修正する

## 1. 計算の流れ(p73) | 入力層、中間層(１つ)、出力層の３層ニューラルネットワークの場合  
➡ 微分の"連鎖律"を使用して誤差勾配を効率よく計算する 

## 誤差勾配の計算の流れ  

#### (1) 誤差関数＝平均二乗誤差  
> ### $E(y) = \dfrac{1}{2} \sum_{j=1}^{J} (y_{j} - d_{j})^{2} = \dfrac{1}{2}||y - d||^{2}$

#### (2) 出力層の活性化関数 = 恒等写像
> ### $y = u^{(L)}$

#### (3) 総入力の計算  
> ### $u^{(l)} = w^{(l)}z^{(l-1)} + b^{(l)}$

### 1.1. 誤差関数(二乗誤差)$E$を$y$について微分する   
> ### ${\frac{\partial E}{\partial y} = \frac{\partial}{\partial y}\frac{1}{2}||y-d||^2 = y - d}$  
>> 目的変数とのズレを見る誤差の計算から活性化関数へ戻していく部分である  

### 1.2. $y_{u}$を$u$について微分する (※今回は恒等関数を活性化関数に使用している(そのまま出力している))   
> ### ${\frac{\partial y(u)}{\partial u} = \frac{\partial u}{\partial u} = 1}$  
>> 活性化関数から出力層への部分である  

### 1.3. 中間層から得られた入力$z$をもとに$w^{(I)}z^{(I-1)} + b^{(I)}$で算出した$u$を$w$について微分する   
> ### ${\frac{\partial u}{\partial w_{ji}}=\frac{\partial}{\partial w_{ji}}(w^{(I)}z^{(I-1)}+b^{(I)})=\frac{\partial}{\partial w_{ji}}
\left(\left[\begin{array}{rrrrr}
w_{11}z_{i} & + ...+ & w_{1i}z_{i} & + ...+ & w_{1I}z_{I} \\
& & \vdots & & \\
w_{j1}z_{i} & + ...+ & w_{ji}z_{i} & + ...+ & w_{jI}z_{I} \\
& & \vdots & & \\
w_{J1}z_{i} & + ...+ & w_{Ji}z_{i} & + ...+ & w_{JI}z_{I} \\
\end{array} \right]
+
\left[\begin{array}{r}
b_{1} \\
\vdots \\
b_{j} \\
\vdots \\
b_{J} \\ \end{array}\right]\right)
=
\left[\begin{array}{r}
0 \\
\vdots \\
z_{j} \\
\vdots \\
0 \\ \end{array} \right]}$  

>> 出力層から入力への部分である。  

### 1.4. 微分の連鎖律   
> 微分の連鎖律によって、(1)~(3)までの計算を掛け合わせることによって、誤差関数$E$を重み$w$で微分した結果を得ることができる。  
> $w$(2)は2層の中間層を示している。
